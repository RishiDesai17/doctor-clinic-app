```python
from sklearn.datasets import load_breast_cancer
breast_cancer_dataset=load_breast_cancer()
```

```python
print(breast_cancer_dataset.DESCR)
```

# So this dataset has 569 rows of data and each row has 30 attributes

```python
breast_cancer_dataset.data.shape
```

# Now we convert it into a pandas dataframe

```python
df=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
df_target=pd.DataFrame(breast_cancer_dataset.target,columns=["Cancer"])
```

```python
df.head()
```

```python
v=pd.concat([df,df_target],axis=1)
v.head()
```
